<h2>Stroke Data Analysis</h2>

<h3>Abstract</h3>

Data for this work was imported from a kaggle.com dataset. The dataset can be accessed with the following link: https://www.kaggle.com/asaumya/healthcare-dataset-stroke-data. The data contains information about patient age, gender, hypertension, heart disease, marital status, work status, residence type, average glucose level, BMI and smoking status all in relation to stroke. In our analysis we were working to find relationships between the various features and stroke prevalence. First, gender was explored with respect to stroke, this yielded an interesting result that the probability of stroke in males is higher, despite a smaller number of males overall in the sample.
Next the effect of hypertension on strokes was explored and again an effect was found that showed an increased probability (from 1.48 in those without hypertension to 4.92 in those with hypertension) of stroke in those with prior hypertension. The rate was more than 3 times higher in those with hypertension than those without. Finally, the effect of BMI on heart disease was investigated as heart disease is a precursor to stroke. Again a relationship was found that indicated that those with higher BMI had a higher rate of heart disease. This analysis was done from the perspective of reducing the rate of stroke as a public health issue. Through using these findings people more at risk of stroke can hopefully be more effectively targeted. 

<h3>Introduction and Findings</h3>

Stroke is one of the many ways in which people die each year. However, it is also one of the largest killers, rating 5th in the USA in 2017 with 142,132 people at an instance of 5.18% of overall causes of death. We undertook this analysis to find actionable insights and to find any relationships that might provide a warning to those at risk of stroke, so they might reduce their risk through lifestyle interventions. Through our analysis, we observed trends and relationships between age, gender, hypertension and BMI and their overarching effect on stroke. The data we collected contained information on 43,400 unique individuals. First hypotheses were built to test relationships, then data was analysed and statistical tests were performed to validate our hypotheses.

Firstly, the data on age was investigated. The rate of hypertension, heart disease and stroke for each decade in age was analyzed and indicated that as age increased, the rate of all three factors increased significantly up to ~70 years on age, then decreased after that. Our assumption is that after age 70 patients may reach the end of their lives and therefore the rate falls for those over 70. The effect of age on stroke and the other realted features can be seen in the figure below: 







This figure clearly indicates that as age increases, the prevalence of stroke, hypertension and heart disease all increase. This shows that older patients are more at risk for cardiovascular problems and therefore must be targeted with interventions. In order to reduce the rate of stroke deaths in the USA, we must begin testing for at risk patients (older) when they are younger, before the risk of stroke increases.

Next, we investigated the effect of gender on stroke, this was again to see if any particular demographic might be prone to higher rates and therefore target them with intervention. We hypothesized a non-negligible effect of gender on stroke. Through analysis, those of male and female gender were separated and the probabilities of stroke given each gender were calculated. This yielded a 1.98% probability in males and a 1.67% probability in females. At first glance this looks like a negligible outcome, however after statistical testing (ANOVA) it was found that there was a significant effect of gender on stroke. This testing also confirmed our alternative hypothesis and verified with a high level of confidence, our rejection of the null hypothesis. An easy visualisation of the effect of gender on stroke can be seen below:



This figure displays the different probabilities for males and females related to stroke. Males have a higher chance of having a stroke compared to females. Therefore males need extra attention for early intervention for prevention of stroke. This is in order to reduce the probability of stroke overall in the wider population. 

Further, we analysed the effect of hypertension on stroke. We divided the population into two groups: those with and those without hyptertension. Within these two groups the probability of having a stroke was calculated using a two  proportion test. The results were that those with hypertension had a probability of 4.92% and those without hypertension had a reduced probability of 1.48%. This indicates that those with hypertension will need extra treatment and early intervention in order to reduce the risk of stroke. The proportion test gave a negligible p-value, which allowed us to reject the null hypothesis with great confidence. Addtionally, Cohen's h (a test for effect size in two proportions) yielded a result of 0.23 which given Cohen's guidelines, is a small effect. This does not however, mean no effect. The fiogure below shows the difference in probability in these two categories:


Finally, we investigated the effect of BMI on heart disease as a predictor for stroke. We again separated our population in to those with heart disease and those without heart disease. We then plotted distributions for each subset and calculated the mean for each. Our null hypothesis was the BMI would have no effect on the rate of heart disease. Using our two distributions we conducted a Welch's t-test, this provided us with an extremely small p-value (1.27e-41) which allowed us to reject the null hypothesis with conviction. In order to test the meaning of these results, Cohen's h was implemented to calculate effect size. This gave a result of 0.25; a small effect given Cohen's recommendations. Additionally, to corroborate our p-value, a power test was conducted and yielded a result of 1, meaning that there was complete certainty that we did not commit a type II error. <h3>Conclusions</h3>

Given these findings we determined that certain patients need specific targeting in order to reduce stroke as a public health issue. Additionally, we think a sample of a smaller size would be more efficacious in determining a more truthful and intelligible result. In reality, we recommend that males, the elder, those with high BMI and those with hypertension need to be given extra attention. Specifically they might be given more frequent check-ups may be prescribed a healthier diet and perhaps some exercise. This can hopefully serve to reduce the incidence of stroke in the general US population.
